## Creating the Dataset

There were some issues regarding the loading of dataset from the original repository so I created another smaller subset. For some reason, creating this dataset via a script leads to failures especially during extraction of the frames. To move forward with this issue I decided to manually run the components of the `create_dataset.py` script by block.

```
# Load out dependencies
import os
import pickle
import random
from pathlib import Path
from shutil import copyfile, move
```

To make it easier to load the file I have added pickle files to define which videos are included. For the adobe240fps dataset we will use the filename **`Adobe240fps.pkl`**. For my case I just need a small subset of the files to train on so I will be using **`smalldataset.pkl`**.

```
# Load Pickle file
filename = 'smalldataset.pkl'
infile = open(filename, 'rb')
read_dict = pickle.load(infile)
infile.close()

testVideoNames = read_dict['testVideoNames']
trainVideoNames = read_dict['trainVideoNames']
```

Here I am just creating the necessary folders for our extract, train, test and validation dataset.

For `root = Path('./data')` you can change `./data` to the folder in your drive where you would want to create the dataset.
Once the cell is run you should be able to see the new root folder as well as the 4 subfolders inside it.

```
root = Path("./data")
if not root.exists():
  os.makedirs("./data")

extractPath      = root/ "extracted"
trainPath        = root/ "train"
testPath         = root/ "test"
validationPath   = root/ "validation"

os.makedirs(extractPath, exist_ok = True)
os.makedirs(trainPath, exist_ok = True)
os.makedirs(testPath, exist_ok = True)
os.makedirs(validationPath, exist_ok = True)
```

For the extraction proper we just have to run the code block below. The ffmpeg script is similar to the one in the original repo but since I am not using the script I have coded the values for the argument directly to the line. You can change the value for the scaling in `scale:<W>:<H>`.

Also, make sure that the directory of your `original_high_fps_videos` is correct. This is what worked based on my drive structure but you should edit it if you had a different naming convention on your end.

```
# For converting the videos to their frames.
# Default scale is 640 x 360 so 360p
# I don't think it would be a good idea to use H:-1 to retain the aspect ratio because it might not match.

failed_list = []
success_list = []
for sublist in [testVideoNames, trainVideoNames]:
#     print("extracting {}".format(item))
    for items in sublist:
        base_dest = Path(extractPath)
        leaf = base_dest/items
        os.makedirs(leaf, exist_ok=True)
        leaf.exists()
#         print("Extracting: {} in {}".format(items, sublist))
        print("Extracting: {}".format(items))
        retn = os.system("ffmpeg -i {} -vf scale=640:360 -qscale:v 2 '{}/%04d.jpg'".format(os.path.join("./unzipfile/original_high_fps_videos",items),leaf))
        if retn:
            failed_list.append(items)
            print("Failed to extract {}".format(items))
        else:
            success_list.append(items)
            print("Extracted {}".format(items))
print('\n*****************************\nFailed to extract: {} \nCompleted extraction: {}'.format(failed_list, success_list))
```

Next up we would be dividing the extracted frames into corresponding subsets and into individual folders with 12 frames each.
There was an issue here during training where I got an `IndexOutofRange` error which I believe was caused by the last folders having less than 12 frames so those frames that cannot be grouped into 12 are dropped.

```
base_pth = Path("./data")
extract_pth = base_pth / "extracted"


for folder in [testVideoNames, trainVideoNames]:
    foldercnt = 0
    for sub_folder in folder:
        images = sorted(os.listdir(extract_pth / sub_folder))
        batch_list = [images[x:x+12] for x in range(0, len(images),12)]      
        batch_list = batch_list[:-1] # Do not include the last folder that would be incomplete
        for item in batch_list:
            if (len(item) % 12 ==0):
                if (folder == trainVideoNames):
                    dummy = base_pth / "train" / str(foldercnt)
                else:
                    dummy = base_pth/ "test" / str(foldercnt)
                os.makedirs(dummy, exist_ok= True)
                for filename in item:
                    src = extract_pth / sub_folder/ filename
                    dst = dummy / filename
                    copyfile(src,dst)
                foldercnt +=1
            else:
                break
# Create a validation subset frame first before moving the items to the folders

base_pth = Path("./data")
val_pth = base_pth /"validation"
images = os.listdir("./data/test")
k = int(0.1*len(images))
to_move = max(1,k)          
```

Once we have moved our testing and training data we can now proceed with moving the subset of folders we want from the test data to the validation data.

Note that the operation here should be move since our validation and test data should be unique.

```
indices = random.sample(range(len(images)), to_move)

for items in indices:
    src = base_pth / 'test' / str(items)
    dst = base_pth / 'validation' / str(items)
    move(src,dst)
    
print('Completed Creating the Test Train Data')
print('Total train: {}\nTotal test: {}\nTotal validation: {}'.format(len(os.listdir(trainPath)),len(os.listdir(testPath)),len(os.listdir(validationPath))))
    
```
